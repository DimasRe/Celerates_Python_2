# LotFrontage

median adalah pilihan yang kuat untuk mengisi nilai yang hilang dalam kolom numerik seperti LotFrontage karena ketahanannya terhadap outliers, kemampuannya untuk merepresentasikan distribusi data yang tidak normal, dan relevansinya sebagai ukuran pemusatan untuk data numerik.

```python
# Mengisi nilai yang hilang di kolom LotFrontage dengan median
df['LotFrontage'].fillna(df['LotFrontage'].median(), inplace=True)
```

```python
# Mengecek apakah masih ada missing values
df['LotFrontage'].isnull().sum()
```
